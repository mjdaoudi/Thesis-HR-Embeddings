```python
def get_average_vector(list, model, word2vec):
    if word2vec:
        t = []
        OOV = 0
        for i in list:
            try:
                t.append(model.wv[i])
            except:
                OOV += 1
                pass
        t = np.mean(np.array(t),axis=0)
        return t
    else:
        t = []
        OOV = 0
        for i in list:
            try:
                t.append(model[i])
            except:
                OOV += 1
                pass
        t = np.mean(np.array(t),axis=0)
        return t
```
