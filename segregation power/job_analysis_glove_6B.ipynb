```python
import pandas as pd
import gensim
import numpy as np
```

```python
desc = pd.read_pickle("data/jobDescription4.pkl")
```

```python
desc["gensim_shape"] = desc["Description_processed"].apply(gensim.utils.simple_preprocess)
```

```python
c_word2vec = gensim.models.Word2Vec.load("models/custom/Word2Vec.model")
c_fasttext = gensim.models.Word2Vec.load("models/custom/FastText.model")

c_glove = gensim.models.KeyedVectors.load_word2vec_format("models/custom/GloVe.txt")
glove = gensim.models.KeyedVectors.load_word2vec_format("models/pre-trained/wvglove.6B.100d.txt")
```

```python
c_word2vec.wv["core"]
```

```python
def get_average_vector(list, model, word2vec):
    if word2vec:
        t = []
        OOV = 0
        for i in list:
            try:
                t.append(model.wv[i])
            except:
                OOV += 1
                pass
        t = np.mean(np.array(t),axis=0)
        return t
    else:
        t = []
        OOV = 0
        for i in list:
            try:
                t.append(model[i])
            except:
                OOV += 1
                pass
        t = np.mean(np.array(t),axis=0)
        return t
```

```python
desc.head(10)
```

```python
desc["avg_vect"] = desc["gensim_shape"].apply(lambda x: get_average_vector(x, glove, False))
```

```python
desc_real = desc[desc["board"]!= "web"][["searched_title", "avg_vect"]]
desc_real
```

```python
umap_test = pd.DataFrame(desc_real["avg_vect"].to_list())
```

```python
from umap import UMAP
import plotly.express as px

umap_2d = UMAP(n_components=2, init='random', random_state=0)
proj_2d = umap_2d.fit_transform(umap_test)
```

```python
proj_2d
```

```python
fig_2d = px.scatter(
    proj_2d, x=0, y=1,
    color=desc_real["searched_title"], labels={'color': 'searched_title'},
    height=800
)
fig_2d.show()
```

```python
umap_3d = UMAP(n_components=3, init='random', random_state=0)
proj_3d = umap_3d.fit_transform(umap_test)
```

```python
fig_3d = px.scatter_3d(
    proj_3d, x=0, y=1, z=2,
    color=desc_real["searched_title"], labels={'color': 'searched_title'},
    height=800
)
fig_3d.update_traces(marker_size=3)
fig_3d.show()
```
